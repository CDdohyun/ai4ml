```python
import pandas as pd

df = pd.read_csv('data/earthquakes.csv')
df.head()
```

```python
df.columns
```

```python
df.rename(columns={'title' : 'location'}, inplace=True)
df.head()
```

```python
df_place = df.place.str.extract(r', (.*$)')  #', ' 이후만 잘라내기
df_place[0].unique()
```

```python
df_location = df.location.str.extract(r', (.*$)')
df_location[0].equals(df_place[0])     #place변수와 title변수가 같은가?
```

```python
df['location'] = df_location
df.head()
```

```python
df.assign(   #여러 변수 추가
    in_nevada = df.location == 'Nevada',   #df[boolean mask]가 아니라 그냥 boolean mask만 적는다
    in_ca = df.location == 'CA',
    in_neither = lambda x: ~x.in_nevada & ~x.in_ca
).head()
```
