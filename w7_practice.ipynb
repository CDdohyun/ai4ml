```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

fb = pd.read_csv(
    'data/fb_stock_prices_2018.csv', index_col='date', parse_dates=True
)
quakes = pd.read_csv('data/earthquakes.csv')
covid = pd.read_csv('data/covid19_cases.csv').assign(
    date=lambda x: pd.to_datetime(x.dateRep, format='%d/%m/%Y')
).set_index('date').replace(
    'United_States_of_America', 'USA'
).sort_index()['2020-01-18':'2020-09-18']
```

```python
fb.head()
```

```python
quakes.head()
```

```python
covid.head()
```

```python
fb.first('1W').plot(
    y=['open', 'high', 'low', 'close'],    #    -> y = [...]
    style=['o-b', '--r', ':k', '.-g'],     #    
    title='Facebook OHLC Prices during 1st Week of Trading 2018'
).autoscale()
```

```python
fb.plot(
    kind='line',
    subplots=True,      #    : subplots=True
    layout=(3, 2),      # grid  (3 2)
    figsize=(15, 10),
    title='Facebook Stock 2018'
)  #     1 
```

```python
new_cases_rolling_avg = covid.pivot_table(
    index=covid.index,
    columns='countriesAndTerritories',
    values='cases'
).rolling(7).mean()

new_cases_rolling_avg.head()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

new_cases_rolling_avg[['China']].plot(kind='line', ax=axes[0], style='-.c')
new_cases_rolling_avg[['Italy', 'Spain']].plot(kind='line', ax=axes[1], style=['-', '--'])
new_cases_rolling_avg[['Brazil', 'India', 'USA']].plot(ax=axes[2], style=['--', ':', '-'])
```

```python
cols = [col for col in new_cases_rolling_avg.columns
        if col not in ['Brazil', 'India', 'USA']]  # 3    = cols

cols2 = [col for col in new_cases_rolling_avg.columns
        if col not in ['Brazil', 'India', 'USA', 'Italy', 'Spain']]  # 5    = cols2

new_cases_rolling_avg.assign(
    Italy_Spain = lambda x: x.Italy + x.Spain
).sort_index(axis=1).assign(                    #Others        .
    Others = lambda x: x[cols2].sum(axis=1)     #5   = Others, axis=1   
).drop(columns=cols).plot(                      #brazil & india & usa (not in cols), Italy_spain & others (assign  )     .
    kind='area', figsize=(15, 5), title='7-days rolling average of new COVID-19 cases\n(source = ECDC)'   # \n =    
)
```

- Change the graph below to stacked bar plot, and normalized stacked bar plot (like below).

```python
indonesia_quakes = quakes.query('parsed_place == "Indonesia"').assign(
    time=lambda x: pd.to_datetime(x.time, unit='ms'),
    earthquake=1   #     1 
).set_index('time').resample('1D').sum()   # .resample('1D') == .groupby(pd.Grouper(freq='1D'))

indonesia_quakes.head()   #sum()    
```

```python
# format the datetimes in the index for the x-axis
indonesia_quakes.index = indonesia_quakes.index.strftime('%b\n%d')

indonesia_quakes.plot(
    y=['earthquake', 'tsunami'], kind='bar', figsize=(15, 3),    #y [...]      
    rot=0, label=['earthquakes', 'tsunamis'], 
    title='Earthquakes and Tsunamis in Indonesia '
          '(September 18, 2018 - October 13, 2018)'
)

# label the axes (discussed in chapter 6)
plt.xlabel('date')
plt.ylabel('count')
```

```python
indonesia_quakes.plot(   #stacked bar plot
    y=['earthquake', 'tsunami'], kind='bar', figsize=(15, 3),    #y [...]      
    rot=0, label=['earthquakes', 'tsunamis'], stacked=True,      #stacked=True
    title='Earthquakes and Tsunamis in Indonesia '
          '(September 18, 2018 - October 13, 2018)'
)
```

```python
normalized_indonesia_quakes = indonesia_quakes.fillna(0)[['tsunami','earthquake']].apply(   #,   normalized  (mag)  
    lambda x: x / x.sum(), axis=1)    # 

normalized_indonesia_quakes.head()
```

```python
normalized_indonesia_quakes.plot(                               #normalized & stacked bar plot
    y=['earthquake', 'tsunami'], kind='bar', figsize=(15, 3),
    rot=0, label=['earthquakes', 'tsunamis'], stacked=True,
    title='Earthquakes and Tsunamis in Indonesia '
          '(September 18, 2018 - October 13, 2018)'
)
```
